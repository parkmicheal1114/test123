# 가변인수 사용시 , 반환시에도 가변인수의 형태와 동일하게 사용해야 함.

>> **kwargs 사용 했으면 **me 이런 형태를 갖춰야 함.

```python
# 실습퀴즈7. 함수
# 나만의 프로필 함수를 만들어 보세요. 
# for kword, arg in kwargs.items():
def my_profile(**kwargs):
    print('저를 소개합니다.')    # 키워드 가변 인수 활용
    for kword , arg in kwargs.items():
         print(f"{kword} : {arg}")

# 여러분의 프로필을 넣어보세요. 
me = {'name':'박기용', 'age':18, 'MBTI':'INFP'} 

my_profile(**me)
```
